## Introduccin a Nave Bayes

En machine learning, los modelos Naive Bayes son una clase especial de algoritmos de **clasificacin**.

Se basan en una tcnica de clasificacin estadstica llamada _Teorema de Bayes_. Naive Bayes **asume que las variables predictoras son independientes entre s**. Y gracias a esto, se puede aplicar la estadstica para calcular la probabilidad posterior de que ocurra un cierto evento _A_, dadas algunas probabilidades de eventos anteriores.

Como su propio nombre indica (Naive=ingenuo) son modelos simples, pero muy potentes en determinadas tareas (por ejemplo, en todo lo relacionado con PLN).


## Recordatorio de conceptos de probabilidad

Simplificamos el concepto de probabilidad haciendo referencia a los sucesos.

Cuando hablamos de un suceso _A_, por ejemplo, obtener un nmero par al realizar un lanzamiento de un dado equilibrado de 6 caras (asumimos que aparecen los nmeros de 1 a 6), podemos considerar que esta probabilidad es:

_P(nmeropar) = P(obtenerun2) + P(obtenerun4) + P(obtenerun6) / 1 = 0,5_

Trivialmente observamos que hemos realizado el cociente entre las probabilidades asociadas a los casos favorables entre la probabilidad asociada al conjunto de posibles resultados (generalmente 1).

**Probabilidad de que ocurra el suceso que queremos: cociente entre las probabilidades asociadas a los casos favorables entre la probabilidad asociada al conjunto de posibles resultados.**

## Teorema de Bayes



Diremos que un suceso _A_ est condicionado por otro suceso _B_ cuando la probabilidad de _A_ se calcula a partir de la confirmacin del suceso _B_.

P (A) : probabilidad de que A sea cierto
P (B): probabilidad de que B sea cierto
P (B / A):  probabilidad de que se de B dado A
P (A / B): probabilidad posterior de que se de A dado B

## Algoritmo Nave Bayes

_Nave Bayes Classifier_ utiliza el teorema de Bayes para predecir las probabilidades de pertenencia para cada clase, como la probabilidad de que un registro o punto de datos dado pertenezca a una clase en particular. La clase con la probabilidad ms alta se considera como la clase ms probable. Esto tambin se conoce como **MAP (Maximum A Posteriori)**.

Este clasificador asume que todas las caractersticas que predicen la variable objetivo son independientes entre s.

Nave Bayes utiliza las probabilidades _a priori_ (conocidas) de los sucesos para estimar la probabilidad de sucesos futuros por medio del TB.

El modelo Naive Bayes es fcil de construir y particularmente til para conjuntos de datos muy grandes. A pesar de la simplicidad, se sabe que Nave Bayes supera incluso a los mtodos de clasificacin altamente sofisticados en determinados escenarios.

_Nota: Sabemos que el supuesto de independencia nunca es correcto, pero a menudo funciona bien en la prctica._

```python
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

df.columns
```

Para hacer un ejemplo completo habra que efectuar el EDA. Eso lo veremos en el laboratorio.

```python
X = df.drop(['income'], axis=1)
y = df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
# Tenemos que transformar los datos para que las libreras pueden hacer su trabajo
categorical = [col for col in X_train.columns if X_train[col].dtypes == 'O']
categorical
```

```python
numerical = [col for col in X_train.columns if X_train[col].dtypes != 'O']
numerical
```

```python
X_train[categorical].head()
```

```python
import category_encoders as ce

encoder = ce.OneHotEncoder(cols=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 
                                 'race', 'sex', 'native_country'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB

# instantiate the model
gnb = GaussianNB()

# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

y_pred
```
