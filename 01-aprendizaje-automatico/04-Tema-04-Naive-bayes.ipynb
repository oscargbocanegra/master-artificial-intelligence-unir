## Introducción a Naïve Bayes

En machine learning, los modelos Naive Bayes son una clase especial de algoritmos de **clasificación**.

Se basan en una técnica de clasificación estadística llamada _Teorema de Bayes_. Naive Bayes **asume que las variables predictoras son independientes entre sí**. Y gracias a esto, se puede aplicar la estadística para calcular la probabilidad ‘posterior’ de que ocurra un cierto evento _A_, dadas algunas probabilidades de eventos ‘anteriores’.

Como su propio nombre indica (Naive=ingenuo) son modelos simples, pero muy potentes en determinadas tareas (por ejemplo, en todo lo relacionado con PLN).


## Recordatorio de conceptos de probabilidad

Simplificamos el concepto de probabilidad haciendo referencia a los sucesos.

Cuando hablamos de un suceso _A_, por ejemplo, obtener un número par al realizar un lanzamiento de un dado equilibrado de 6 caras (asumimos que aparecen los números de 1 a 6), podemos considerar que esta probabilidad es:

_P(número par) = P(obtener un 2) + P(obtener un 4) + P(obtener un 6) / 1 = 0,5_

Trivialmente observamos que hemos realizado el cociente entre las probabilidades asociadas a los casos favorables entre la probabilidad asociada al conjunto de posibles resultados (generalmente 1).

**Probabilidad de que ocurra el suceso que queremos: cociente entre las probabilidades asociadas a los casos favorables entre la probabilidad asociada al conjunto de posibles resultados.**

## Teorema de Bayes



Diremos que un suceso _A_ está condicionado por otro suceso _B_ cuando la probabilidad de _A_ se calcula a partir de la confirmación del suceso _B_.

P (A) : probabilidad de que A sea cierto
P (B): probabilidad de que B sea cierto
P (B / A):  probabilidad de que se de B dado A
P (A / B): probabilidad posterior de que se de A dado B

## Algoritmo Naïve Bayes

_Naïve Bayes Classifier_ utiliza el teorema de Bayes para predecir las probabilidades de pertenencia para cada clase, como la probabilidad de que un registro o punto de datos dado pertenezca a una clase en particular. La clase con la probabilidad más alta se considera como la clase más probable. Esto también se conoce como **MAP (Maximum A Posteriori)**.

Este clasificador asume que todas las características que predicen la variable objetivo son independientes entre sí.

Naïve Bayes utiliza las probabilidades _a priori_ (conocidas) de los sucesos para estimar la probabilidad de sucesos futuros por medio del TB.

El modelo Naive Bayes es fácil de construir y particularmente útil para conjuntos de datos muy grandes. A pesar de la simplicidad, se sabe que Naïve Bayes supera incluso a los métodos de clasificación altamente sofisticados en determinados escenarios.

_Nota: Sabemos que el supuesto de independencia nunca es correcto, pero a menudo funciona bien en la práctica._

```python
mainpath = "/proyectos/Master-IA-Unir/02-aprendizaje-automatico/datasets/"
filename = "adult.csv"
fullpath = os.path.join(mainpath, filename)
data = pd.read_csv(fullpath)
```

```python
# dimensiones del dataset

df.shape
```

```python
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

df.columns
```

```python
df.info()
```

Para hacer un ejemplo completo habría que efectuar el EDA. Eso lo veremos en el laboratorio.

```python
X = df.drop(['income'], axis=1)
y = df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
# Tenemos que transformar los datos para que las librerías pueden hacer su trabajo
categorical = [col for col in X_train.columns if X_train[col].dtypes == 'O']
categorical
```

```python
numerical = [col for col in X_train.columns if X_train[col].dtypes != 'O']
numerical
```

```python
X_train[categorical].head()
```

```python
# Install the category_encoders package
#%pip install -U category_encoders

import category_encoders as ce

encoder = ce.OneHotEncoder(cols=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 
                                 'race', 'sex', 'native_country'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB

# instantiate the model
gnb = GaussianNB()

# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

y_pred
```

```python
y_test
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
(accuracy)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
# Mostrar precision and recall, junto con el resto de métricas
print(classification_report(y_test, y_pred, digits=3))
```

```python
precision = metrics.precision_score(y_test, y_pred, pos_label=' <=50K')
recall = metrics.recall_score(y_test, y_pred, pos_label=' <=50K')
f1 = metrics.f1_score(y_test, y_pred, pos_label=' <=50K')
```

```python
# Calculate predicted probabilities
y_pred_proba = gnb.predict_proba(X_test)[:, 1]

# Usar la etiqueta con el espacio inicial
metrics.roc_curve(y_test, y_pred_proba, pos_label=' <=50K')
```
